# 全國逐年出生率

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

取出所有的資料

```python
csvdata = pd.read_csv('全國逐年出生率.csv', encoding='utf-8', thousands=',')
csvdata
```

產出折線圖

```python
plt.plot(csvdata['Year'], csvdata['Crude Birth Rate (‰)'], label='Taiwan')

plt.legend()
plt.title('Crude Birth Rate by Years')
plt.xlabel('Year')
plt.ylabel('Crude Birth Rate (‰)')
plt.show()
```
